```python
import pandas as pd

def estimate_prior_likelihood_evidence(data, car_column, model_column):

    prior_probability = data[model_column].value_counts(normalize=True).to_dict()

    grouped_data = data.groupby([car_column, model_column]).size().div(len(data)).to_dict()
    likelihood = {(car, model): prob for (car, model), prob in grouped_data.items()}

    evidence = data[car_column].value_counts(normalize=True).to_dict()

    return prior_probability, likelihood, evidence

def bayes_theorem(prior, likelihood, evidence, car, model):

    posterior = (likelihood.get((car, model), 0) * prior[model]) / evidence[car]
    return posterior


car_column = 'Car'
model_column = 'Model'


specific_car = 'Ford'
specific_model = 'Fiesta'

prior_probability, likelihood, evidence = estimate_prior_likelihood_evidence(Data, car_column, model_column)

posterior_probability = bayes_theorem(prior_probability, likelihood, evidence, specific_car, specific_model)

print(f"Prior Probability for Model {specific_model}: {prior_probability.get(specific_model, 0)}")
print(f"Likelihood for Model {specific_model} given Car {specific_car}: {likelihood.get((specific_car, specific_model), 0)}")
print(f"Evidence for Car {specific_car}: {evidence.get(specific_car, 0)}")
print(f"Posterior Probability for Model {specific_model} given Car {specific_car}: {posterior_probability}")
```
