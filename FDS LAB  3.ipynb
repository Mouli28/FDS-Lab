lets use the car data set and plot a scatter plot using matplotlib for car and co2 column rotation 45 is given to tilt the car names 45 degree to avoid overlapping

```python
plt.xticks(rotation=45)
plt.plot(Data.Car,Data.CO2,marker='o')
```

Subplot technique is used to plot two different graphs in one display

```python
x=Data.Car
y=Data.CO2

plt.subplot(1,2,1)
plt.xticks(rotation=45)
plt.plot(x,y)



x=Data.Car
y=Data.Volume

plt.subplot(1,2,2)
plt.xticks(rotation=45)
plt.plot(x,y)
```
