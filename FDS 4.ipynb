Types of sampling

Random sampling (normal sampling)

```python
from random import sample
print(Data.sample(n=2))
```

Stratified sampling(sampling after grouping the datas into groups based on some criteria)

here sampling is done if the number of element in a group is more than s value else the group will be printed as it is cause the size of the code is already low.

```python
s = 2
print(Data.groupby('Car', group_keys=False).apply(lambda x: x.sample(n=s) if len(x) >= s else x))
```

Systematic sampling(where datas are extracted based on some criteria)

here data from every fifth row is taken and here iloc is a pandas method that is used to index  

```python
print(Data.iloc[::5])
```
